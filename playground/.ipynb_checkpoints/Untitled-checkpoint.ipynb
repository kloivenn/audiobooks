```python
import re
```

```python
with open('01_start.txt', 'r') as file:
    data = file.read()
    
data
```

```python
with open("dict_example.txt", "r") as file:
    dic = file.readlines()
    
repl = list(map(lambda el: el.replace('\n', '').split('/'), dic))
for r in repl:
    if r[2] == 'b':
        r[0] = '(\W)' + r[0] + '(\W)'
        r[1] = '\g<1>' + r[1] + '\g<2>'
    if r[2] == 'r':
        r[0] = r[0] + '(\W)'
        r[1] = r[1] + '\g<1>'
    if r[2] == 'l':
        r[0] = '(\W)' + r[0]
        r[1] = '\g<1>' + r[1]
        
repl
```

```python
with open("dict_example.txt", "r") as file:
    dic = file.readlines()
    
repl = list(map(lambda el: el.replace('\n', '').split('/'), dic))
for r in repl:
    if r[2] == 'b':
        r[0] = '\\b' + r[0] + '\\b'
    if r[2] == 'r':
        r[0] = r[0] + '\\b'
    if r[2] == 'l':
        r[0] = '\\b' + r[0]
        
repl
```

```python
newData = data

for r in repl:
    newData = re.sub(r[0], r[1], newData)
    
newData
```
