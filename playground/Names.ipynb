```python
import re
```

```python
with open('../text/fanfics.txt', 'r') as file:
    data = file.read()
    
data
```

```python
with open("../dicts/00_names_hom.txt", "r") as file:
    dic = file.readlines()
    
repl = list(map(lambda el: el.replace('\n', '').split('/s/'), dic))
for r in repl:
    if r[2] == 'b':
        r[0] = '\\b' + r[0] + '\\b'
    if r[2] == 'r':
        r[0] = r[0] + '\\b'
    if r[2] == 'l':
        r[0] = '\\b' + r[0]
    if int(r[3]) == 0:
        fl = re.IGNORECASE
    else:
        fl = 0

    newText = re.sub(r[0], r[1], data, flags = fl)
        
newText
```

```python
with open('text.txt', 'w') as out:
    out.write(newText)
```
